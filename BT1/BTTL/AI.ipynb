```python
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
tong = a + b
hieu = a - b
tich = a*b
thuong = a/b
print("Tổng: ", tong)
print("Hiệu: ", hieu)
print("Tích: ", tich)
print("Thương: ", thuong)
```

```python
chuoi = " Hello World "
print(f"Gốc: '{chuoi}'")
cutted = chuoi.strip()
print(f"Đã cắt: '{cutted}'")
hoa = cutted.upper()
print(f"Hoa: '{hoa}'")
thuong = cutted.lower()
print(f"Thường: '{thuong}'")
replaced_string = cutted.replace('H', 'J')
print(f"Thay 'H' bằng 'J': '{replaced_string}'")
```
