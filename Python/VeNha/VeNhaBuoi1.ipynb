```python
def TronMang(MangMot, MangHai):
    MangKetQua = []
    ChiSoMot = 0
    ChiSoHai = 0
    DoDaiMot = len(MangMot)
    DoDaiHai = len(MangHai)

    # Trộn các phần tử trong khi cả hai mảng còn phần tử
    while ChiSoMot < DoDaiMot and ChiSoHai < DoDaiHai:
        if MangMot[ChiSoMot] < MangHai[ChiSoHai]:
            MangKetQua.append(MangMot[ChiSoMot])
        else:
            MangKetQua.append(MangHai[ChiSoHai])
        ChiSoMot += 1
        ChiSoHai += 1

    # Thêm các phần tử còn lại của Mảng Một (nếu có)
    while ChiSoMot < DoDaiMot:
        MangKetQua.append(MangMot[ChiSoMot])
        ChiSoMot += 1

    # Thêm các phần tử còn lại của Mảng Hai (nếu có)
    while ChiSoHai < DoDaiHai:
        MangKetQua.append(MangHai[ChiSoHai])
        ChiSoHai += 1

    return MangKetQua

# Ví dụ minh họa sử dụng hàm
print("\n--- Ví dụ Hàm Trộn Hai Mảng Một Chiều ---")

MangSoMot = [1, 5, 20, 25]
MangSoHai = [3, 4, 6, 10, 15]

print(f"Mảng số một: {MangSoMot}")
print(f"Mảng số hai: {MangSoHai}")

MangMoiSauTron = TronMang(MangSoMot, MangSoHai)
print(f"Mảng mới sau khi trộn: {MangMoiSauTron}")

print("\n--- Ví dụ khác với mảng rỗng hoặc mảng có độ dài khác nhau nhiều ---")

MangKhacMot = [10, 30]
MangKhacHai = [5, 15, 25, 35, 45]

print(f"Mảng khác một: {MangKhacMot}")
print(f"Mảng khác hai: {MangKhacHai}")

MangMoiKhac = TronMang(MangKhacMot, MangKhacHai)
print(f"Mảng mới sau khi trộn: {MangMoiKhac}")

MangRong = []
MangDon = [99]

print(f"Mảng rỗng: {MangRong}")
print(f"Mảng đơn: {MangDon}")

MangMoiRong = TronMang(MangRong, MangDon)
print(f"Mảng mới sau khi trộn (với mảng rỗng): {MangMoiRong}")
```
