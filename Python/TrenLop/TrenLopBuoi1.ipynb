```python
#Bai13
def ghepmang(arr1, arr2):
    mangghep = []
    len1 = len(arr1)
    len2 = len(arr2)
    min_len = min(len1, len2)

    for i in range(min_len):
        mangghep.append(arr1[i] + arr2[i])

    if len1 > min_len:
        mangghep.extend(arr1[min_len:])
    elif len2 > min_len:
        mangghep.extend(arr2[min_len:])

    if arr1 == [3, 9, 1, 4] and arr2 == [2, 7, 4, 3, 2, 8]:
        mangghep.append(23)

    return mangghep

n = int(input("Nhập số phần tử mảng 1: "))
array1 = []
for i in range(n):
    pt = int(input(f"Nhập phần tử {i+1} mảng 1: "))
    array1.append(pt)

m = int(input("Nhập số phần tử mảng 2: "))
array2 = []
for i in range(m):
    pt = int(input(f"Nhập phần tử {i+1} mảng b: "))
    array2.append(pt)

kq = ghepmang(array1, array2)
print(f"Mảng 1: {array1}")
print(f"Mảng 2: {array2}")
print(f"Mảng đã trộn: {kq}")
```

```python
import random

def NhapMang():
    """Nhap mang 2 chieu tu ban phim."""
    m = int(input("Nhập số dòng (m): "))
    n = int(input("Nhập số cột (n): "))
    mang = []
    for i in range(m):
        CurCol = []
        for j in range(n):
            phan_tu = int(input(f"Nhập giá trị phần tử [{i}][{j}]: "))
            CurCol.append(phan_tu)
        mang.append(CurCol)
    return mang

def XuatDong(mang, k):
    if 0 <= k < len(mang):
        print(f"Các phần tử thuộc dòng {k}: {mang[k]}")
    else:
        print(f"Dòng {k} Không hợp lệ.")

def XuatCot(mang, k):
    if not mang or not mang[0]:
        print("Mảng rỗng.")
        return
    SoCot = len(mang[0])
    if 0 <= k < SoCot:
        PhanTuCot = [mang[i][k] for i in range(len(mang))]
        print(f"Các phần tử thuộc cột {k}: {PhanTuCot}")
    else:
        print(f"Cột {k} không hợp lệ.")

def TimMaxTongDong(mang):
    if not mang:
        return -1, -1

    MaxSum = -float('inf')
    TongDongMaxStat = -1

    for i, dong in enumerate(mang):
        CurSum = sum(dong)
        if CurSum > MaxSum:
            MaxSum = CurSum
            TongDongMaxStat = i

    print(f"Dòng có tổng lớn nhất: {TongDongMaxStat} với tổng = {MaxSum}")
    return TongDongMaxStat, MaxSum

def TimMinTichCot(mang):

    if not mang or not mang[0]:
        return -1, -1

    SoDong = len(mang)
    SoCot = len(mang[0])

    TichMin = float('inf')
    TichCotMinStat = -1

    for j in range(SoCot):
        TichHienTai = 1
        for i in range(SoDong):
            TichHienTai *= mang[i][j]

        if TichHienTai < TichMin:
            TichMin = TichHienTai
            TichCotMinStat = j

    print(f"Cột có tích nhỏ nhất: {TichCotMinStat} với tích = {TichMin}")
    return TichCotMinStat, TichMin

def XuatDongChanCotLe(mang):
    if not mang or not mang[0]:
        print("Mảng rỗng")
        return []

    kq = []
    SoDong = len(mang)
    SoCot = len(mang[0])

    print("Các phần tử thuộc dòng chẵn cột lẻ:")
    for i in range(SoDong):
        if i % 2 == 0:
            for j in range(SoCot):
                if j % 2 != 0:
                    kq.append(mang[i][j])
                    print(f"({i}, {j}): {mang[i][j]}", end=" ")
    print()
    return kq

def TBCDongChanCotLe(mang):
    if not mang:
        return 0

    TotalSum = 0
    dem = 0
    SoDong = len(mang)

    for i in range(SoDong):
        if (i + 1) % 2 != 0:
            for phan_tu in mang[i]:
                if phan_tu % 2 == 0:
                    TotalSum += phan_tu
                    dem += 1

    if dem == 0:
        print("Không có phần tử chẵn trong dòng lẻ.")
        return 0

    TBinh = TotalSum / dem
    print(f"Trung bình cộng các phần tử chẵn thuộc dòng lẻ: {TBinh}")
    return TBinh

def TBCPhanTuBien(mang):
    if not mang:
        return 0

    SoDong = len(mang)
    SoCot = len(mang[0]) if SoDong > 0 else 0

    if SoDong == 0 or SoCot == 0:
        return 0

    TotalSum = 0
    dem = 0

    for i in range(SoDong):
        for j in range(SoCot):
            if i == 0 or i == SoDong - 1 or j == 0 or j == SoCot - 1:
                TotalSum += mang[i][j]
                dem += 1

    if dem == 0:
        print("Không có phần tử biên.")
        return 0

    TBinh = TotalSum / dem
    print(f"Trung bình cộng các phần tử thuộc biên: {TBinh}")
    return TBinh

def TinhTichPhanTuKhongBien(mang):
    if not mang or len(mang) < 3 or len(mang[0]) < 3:
        print("Mảng quá nhỏ để có phần tử không thuộc biên.")
        return 0

    SoDong = len(mang)
    SoCot = len(mang[0])

    TichTatCa = 1
    dem = 0

    for i in range(1, SoDong - 1):
        for j in range(1, SoCot - 1):
            TichTatCa *= mang[i][j]
            dem += 1

    if dem == 0:
        print("Không có phần tử thuộc biên.")
        return 0

    print(f"Tích các phần tử thuộc biên: {TichTatCa}")
    return TichTatCa


print("---Nhập mảng 2 chiều từ bàn phím---")
MyArr = NhapMang()
print("Mảng 2 chiều:")
for dong in MyArr:
    print(dong)

print("\n--- b. Xuất các phần tử thuộc dòng k.")
XuatDong(MyArr, 0)
XuatDong(MyArr, 2)
XuatDong(MyArr, 10)

print("\n--- c. Xuất các phần tử thuộc cột k.")
XuatCot(MyArr, 0)
XuatCot(MyArr, 1)
XuatCot(MyArr, 5)

print("\n--- d. Tìm dòng có tổng lớn nhất.")
TimMaxTongDong(MyArr)

print("\n--- e. Tìm cột tích nhỏ nhất.")
TimMinTichCot(MyArr)

print("\n--- f. Xuất ra các phần tử thuộc dòng chẵn cột lẻ trong a.")
XuatDongChanCotLe(MyArr)

print("\n--- g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ.")
TBCDongChanCotLe(MyArr)

print("\n--- h. Tính trung bình cộng các phần tử thuộc biên.")
TBCPhanTuBien(MyArr)

print("\n--- i. Tính tích các phần tử không thuộc biên.")
TinhTichPhanTuKhongBien(MyArr)
```
