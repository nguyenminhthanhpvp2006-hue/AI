```python
import heapq
import math

graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'D': 2, 'E': 5},
    'C': {'A': 4, 'F': 2},
    'D': {'B': 2, 'G': 1},
    'E': {'B': 5, 'H': 3},
    'F': {'C': 2, 'I': 3},
    'G': {'D': 1, 'J': 1},
    'H': {'E': 3, 'K': 2},
    'I': {'F': 3, 'L': 1},
    'J': {'G': 1, 'M': 4},
    'K': {'H': 2, 'N': 1},
    'L': {'I': 1, 'M': 2},
    'M': {'J': 4, 'L': 2, 'N': 1},
    'N': {'K': 1, 'M': 1}
}

coordinates = {
    'A': (0, 0), 'B': (1, 1), 'C': (0, 4),
    'D': (2, 0), 'E': (2, 5), 'F': (1, 3),
    'G': (3, 1), 'H': (3, 6), 'I': (2, 2),
    'J': (4, 0), 'K': (4, 7), 'L': (3, 3),
    'M': (5, 2), 'N': (5, 8)
}

def heuristic(node, goal, coordinates):
    if node not in coordinates or goal not in coordinates:
        return float('inf')
    x1, y1 = coordinates[node]
    x2, y2 = coordinates[goal]
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def TimKiemAStar(graph, start, goal, heuristic_func, coordinates):
    priority_queue = [(0 + heuristic_func(start, goal, coordinates), 0, start, [start])]

    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0

    f_score = {node: float('inf') for node in graph}
    f_score[start] = heuristic_func(start, goal, coordinates)

    closed_set = set()

    while priority_queue:
        f_score_hien_Tai, g_score_hien_tai, NodeHienTai, path = heapq.heappop(priority_queue)

        if NodeHienTai == goal:
            return path, g_score_hien_tai

        if NodeHienTai in closed_set:
            continue

        closed_set.add(NodeHienTai)

        for neighbor, weight in graph.get(NodeHienTai, {}).items():
            g_score_du_kien = g_score_hien_tai + weight

            if g_score_du_kien < g_score[neighbor]:
                g_score[neighbor] = g_score_du_kien
                f_score[neighbor] = g_score_du_kien + heuristic_func(neighbor, goal, coordinates)
                new_path = path + [neighbor]
                heapq.heappush(priority_queue, (f_score[neighbor], g_score_du_kien, neighbor, new_path))

    return None, float('inf')

NodeHopLe = list(graph.keys())

print("Các đỉnh hợp lệ là: " + ", ".join(NodeHopLe) + ".")

while True:
    NodeBatDau = input("Nhập đỉnh bắt đầu: ").upper()
    if NodeBatDau in NodeHopLe:
        break
    else:
        print("Đỉnh bắt đầu không hợp lệ.")

while True:
    NodeDich = input("Nhập đỉnh kết thúc: ").upper()
    if NodeDich in NodeHopLe:
        break
    else:
        print("Đỉnh kết thúc không hợp lệ.")

path, cost = TimKiemAStar(graph, NodeBatDau, NodeDich, heuristic, coordinates)

if path:
    print(f"\nĐường đi ngắn nhất từ {NodeBatDau} đến {NodeDich}: {path}")
    print(f"Tổng chi phí: {cost:.2f}")
else:
    print(f"\nKhông tìm thấy đường đi từ {NodeBatDau} đến {NodeDich}.")
```
