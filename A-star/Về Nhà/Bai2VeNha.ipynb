```python
import heapq
import math

def TinhHeuristic(TPHienTai, DaTham, SoTP, dist_matrix, NodeBatDau):
    if DaTham == (1 << SoTP) - 1:
        return dist_matrix[TPHienTai][NodeBatDau]

    min_edge = float('inf')

    for TPTiepTheo in range(SoTP):
        if not (DaTham & (1 << TPTiepTheo)):
            min_edge = min(min_edge, dist_matrix[TPHienTai][TPTiepTheo])
    return min_edge

def tsp_as(dist_matrix):
    SoTP = len(dist_matrix)
    NodeBatDau = 0


    initial_h = TinhHeuristic(NodeBatDau, 1 << NodeBatDau, SoTP, dist_matrix, NodeBatDau)
    pq = [(initial_h, 0, NodeBatDau, 1 << NodeBatDau, [NodeBatDau])]


    MinGCost = {}
    MinGCost[(NodeBatDau, 1 << NodeBatDau)] = 0

    ChiPhiDuongDiMin = float('inf')
    DuongTotNhat = []

    while pq:
        FCost, GCost, NodeHienTai, DaTham, path = heapq.heappop(pq)


        if GCost > MinGCost.get((NodeHienTai, DaTham), float('inf')):
            continue


        if DaTham == (1 << SoTP) - 1:

            CostQuayVe = dist_matrix[NodeHienTai][NodeBatDau]
            TongChiPhiDuongDi = GCost + CostQuayVe
            if TongChiPhiDuongDi < ChiPhiDuongDiMin:
                ChiPhiDuongDiMin = TongChiPhiDuongDi
                DuongTotNhat = path + [NodeBatDau]
            continue


        for next_node in range(SoTP):
            if not (DaTham & (1 << next_node)):
                new_GCost = GCost + dist_matrix[NodeHienTai][next_node]
                new_DaTham = DaTham | (1 << next_node)


                if new_GCost < MinGCost.get((next_node, new_DaTham), float('inf')):
                    MinGCost[(next_node, new_DaTham)] = new_GCost
                    new_path = path + [next_node]

                    new_HCost = TinhHeuristic(next_node, new_DaTham, SoTP, dist_matrix, NodeBatDau)
                    new_FCost = new_GCost + new_HCost
                    heapq.heappush(pq, (new_FCost, new_GCost, next_node, new_DaTham, new_path))

    return ChiPhiDuongDiMin, DuongTotNhat


while True:
    try:
        SoTP_str = input("Nhập số lượng thành phố: ")
        SoTP = int(SoTP_str)
        if SoTP < 2:
            print("Số lượng thành phố phải ít nhất là 2.")
        else:
            break
    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")

dist_matrix = [[0] * SoTP for _ in range(SoTP)]

print("\nNhập ma trận khoảng cách:")
print("Khoảng cách từ thành phố i đến thành phố j. Nhập inf nếu không có đường đi trực tiếp.")

for i in range(SoTP):
    for j in range(SoTP):
        if i == j:
            dist_matrix[i][j] = 0
            continue
        while True:
            try:
                value_str = input(f"Khoảng cách từ thành phố {i} đến thành phố {j}: ")
                if value_str.lower() == 'inf':
                    dist_matrix[i][j] = float('inf')
                else:
                    value = float(value_str)
                    if value < 0:
                        print("Khoảng cách không thể là số âm.")
                    else:
                        dist_matrix[i][j] = value
                        break
            except ValueError:
                print("Đầu vào không hợp lệ.")

print("\nMa trận khoảng cách đã nhập:")
for row in dist_matrix:
    print([round(x, 2) if x != float('inf') else 'inf' for x in row])

min_cost, path = tsp_as(dist_matrix)

if path:
    print(f"\nChi phí tối ưu của chuyến đi: {min_cost}")

    print(f"Lộ trình tối ưu: {' -> '.join(map(str, path))}")
else:
    print("\nKhông tìm thấy chuyến đi hợp lệ.")
```
