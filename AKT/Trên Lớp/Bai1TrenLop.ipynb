```python
#Bai1
import copy
from heapq import heappush, heappop

n = 4
rows = [1,0,-1,0]
cols = [0,-1,0,1]


class priorityQueue:
  def __init__(self):
      self.heap = []

  def push(self, key):
      heappush(self.heap, key)

  def pop(self):
      return heappop(self.heap)

  def empty(self):
    return not self.heap

class nodes:

  def __init__(self, parents, mats, empty_tile_posi, costs, levels):
    self.parents  = parents
    self.mats = mats
    self.empty_tile_posi = empty_tile_posi
    self.costs = costs
    self.levels = levels

  def __lt__(self, nxt):
    return self.costs < nxt.costs

  @staticmethod
  def calculateCosts(mats, final) -> int:
    count = 0
    for i in range(n):
      for j in range(n):
        if mats[i][j] != 0 and mats[i][j] != final[i][j]:
          count += 1
    return count

  @staticmethod
  def newNodes(mats, empty_tile_posi, new_empty_tile_posi,
               levels, parent, final) -> 'nodes':
    new_mats = copy.deepcopy(mats)
    x1 = empty_tile_posi[0]
    y1 = empty_tile_posi[1]
    x2 = new_empty_tile_posi[0]
    y2 = new_empty_tile_posi[1]
    new_mats[x1][y1], new_mats[x2][y2] = new_mats[x2][y2], new_mats[x1][y1]

    costs = nodes.calculateCosts(new_mats, final)
    new_node = nodes(parent, new_mats, new_empty_tile_posi,
                      costs, levels)
    return new_node

def printMatsrix(mats):
  for i in range(n):
    for j in range(n):
      print("%2d " % (mats[i][j]), end = " ")
    print()

def isSafe(x, y):
  return x >= 0 and x < n and y >= 0 and y < n

def printPath(root):
  if root is None:
    return
  printPath(root.parents)
  printMatsrix(root.mats)
  print()

def solve(initial, empty_tile_posi, final):
  pq = priorityQueue()
  costs = nodes.calculateCosts(initial, final)
  root = nodes(None, initial,
               empty_tile_posi, costs, 0)
  pq.push(root)

  while not pq.empty():
    minimum = pq.pop()
    if minimum.costs == 0:
      printPath(minimum)
      return

    for i in range(4):
      new_tile_posi = [
          minimum.empty_tile_posi[0] + rows[i],
          minimum.empty_tile_posi[1] + cols[i]
      ]
      if isSafe(new_tile_posi[0], new_tile_posi[1]):

        child = nodes.newNodes(minimum.mats,
                               minimum.empty_tile_posi,
                               new_tile_posi,
                               minimum.levels + 1,
                               minimum, final)
        pq.push(child)

def get_matrix_input(prompt_message):
  print(prompt_message)
  matrix = []
  for i in range(n):
    while True:
      try:
        row_str = input(f"Nhập hàng {i+1} (các số cách nhau bằng dấu cách, 0 là ô trống): ")
        row = list(map(int, row_str.split()))
        if len(row) == n and all(0 <= x <= n*n - 1 for x in row):
          matrix.append(row)
          break
        else:
          print(f"Tham số không hợp lệ, nhập {n} số trong phạm vi từ 0 đến {n*n - 1}.")
      except ValueError:
        print("Tham số không hợp lệ, các tham số ngăn cách bằng dấu cách")
  return matrix

def find_empty_tile(matrix):
  for r in range(n):
    for c in range(n):
      if matrix[r][c] == 0:
        return [r, c]
  return None

initial_15_puzzle = get_matrix_input("Nhập trạng thái ban đầu của bài toán 15-puzzle:")
empty_tile_posi_15_puzzle = find_empty_tile(initial_15_puzzle)

final_15_puzzle = get_matrix_input("Nhập trạng thái đích của 15-puzzle:")


print("Cách giải:")
solve(initial_15_puzzle, empty_tile_posi_15_puzzle, final_15_puzzle)
```
